## Consistent Compounders

```python
import pandas as pd
import matplotlib.pyplot as plt

index = ['2016', '2017', '2018', '2019', '2020', '2021']
schema = ['revenue', 'pat', 'operating cashflows', 'investing cashflows']
```

<h4>LTTS</h4>

```python
ltts_data = [
    [30662, 4186, 5246, -1540],
    [32483, 4249, 3981, -1767],
    [37471, 5060, 3396, -1656],
    [50783, 7656, 7367, -4861],
    [56191, 8186, 6643, -2805],
    [54497, 6663, 12671, -9646]
]
ltts = pd.DataFrame(data=ltts_data, index=index, columns=schema)
```

```python
# creating the revenue plot
ltts.reset_index().plot(x='index', y='revenue', kind='line', marker='o', figsize = (6, 4))
for row in ltts['revenue']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
lttsDiff = round(ltts['revenue'].pct_change() * 100, 2)
display(lttsDiff)
# creating the revenue growth plot
lttsDiff.reset_index().plot(x='index', y='revenue', marker='o', kind='line', figsize = (6, 4))
for row in lttsDiff:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Revenue Growth")
plt.show()
```

```python
# creating the operating cashflows plot
ltts.reset_index().plot(x='index', y='operating cashflows', kind='line', marker='o', figsize = (6, 4))
for row in ltts['operating cashflows']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Opearating Cashflows")
plt.show()
```

```python
# creating the free cashflows plot
ltts['free cashflows'] = ltts['operating cashflows'] + ltts['investing cashflows']
display(ltts['free cashflows'])

ltts.reset_index().plot(x='index', y='free cashflows', kind='line', marker='o', figsize = (6, 4))
for row in ltts['free cashflows']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Free Cashflows")
plt.show()
```

```python
# creating the free cashflows plot
ltts['operating/pat'] = round(ltts['operating cashflows'] / ltts['pat'], 3)
display(ltts['operating/pat'])

ltts.reset_index().plot(x='index', y='operating/pat', kind='line', marker='o', figsize = (6, 4))
for row in ltts['operating/pat']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Free Cashflows")
plt.show()
```
