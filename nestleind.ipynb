```python
import pandas as pd
import matplotlib.pyplot as plt

index = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
schema = ['revenue', 'pat', 'eps', 'operating cashflows', 'investing cashflows']
```

```python
data = [
  [98063, 11847, 122.9, 16440, -4317.0],
  [81233, 5633, 58.4, 10981, -704.8],
  [94096, 10014, 103.9, 14659, -1259.0],
  [101351, 12252, 127.1, 18178, -1305.6],
  [112162, 16069, 166.7, 20525, -524.1],
  [122953, 19684, 204.2, 22953, 829.9],
  [132902, 20824, 216.0, 24545, -3214.6],
  [146337, 21449, 222.5, 22714, -19570.4]
]
df = pd.DataFrame(data=data, index=index, columns=schema)
```

```python
# creating the eps plot
df.reset_index().plot(x='index', y='eps', kind='line', marker='o', color='red', figsize = (6, 4))
for row in df['eps']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("EPS")
plt.show()
```

```python
# creating the revenue plot
df.reset_index().plot(x='index', y='revenue', kind='line', marker='o', color='red', figsize = (6, 4))
for row in df['revenue']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
diff = round(df['revenue'].pct_change() * 100, 2)
display(diff)

# creating the revenue growth plot
diff.reset_index().plot(x='index', y='revenue', marker='o', kind='line', color='red', figsize = (6, 4))
for row in diff:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Revenue Growth")
plt.show()
```

```python
# creating the operating cashflows plot
df.reset_index().plot(x='index', y='operating cashflows', kind='line', marker='o', color='red', figsize = (6, 4))
for row in df['operating cashflows']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Opearating Cashflows")
plt.show()
```

```python
# creating the free cashflows plot
df['free cashflows'] = df['operating cashflows'] + df['investing cashflows']
display(df['free cashflows'])

df.reset_index().plot(x='index', y='free cashflows', kind='line', marker='o', color='red', figsize = (6, 4))
for row in df['free cashflows']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Free Cashflows")
plt.show()
```

```python
# creating the free cashflows plot
df['operating/pat'] = round(df['operating cashflows'] / df['pat'], 3)
display(df['operating/pat'])

df.reset_index().plot(x='index', y='operating/pat', kind='line', marker='o', color='red', figsize = (6, 4))
for row in df['operating/pat']:
    plt.axhline(y=row, color='grey', linestyle='dashed')
plt.xlabel("Year")
plt.ylabel("Free Cashflows")
plt.show()
```
